# **** Relational Database****

```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
# SQLLite engine has been used
engine = create_engine('sqlite:///Northwind.sqllite')
# Loaded csv file
# I have choosen this process because dont have any database details to make the connection.
data = pd.read_csv('diabetes_database.csv')
```

# 2 TABLE CREATED THROUGH READING CSV FILE

```python
# converting data to sql through sql lite engine and creating 2 tables with single diabetes csv file/
data_10 = data.head(10)
data_10.to_sql('diabetes_selected_database_table', engine)


data.to_sql('diabetes_full_database_table',engine)
```

```python
# After converting this csv data into table now we can use sql query to fetch any details. 
# There are 2 way to execute the query.
# This is the first way to execute this.
pd.read_sql_query('SELECT * from diabetes_full_database_table where Glucose=148', engine)
```

```python
# This is an another way to make the connectiom and fetch the data.
with engine.connect() as con:
    rs=con.execute("SELECT * from diabetes_selected_database_table")
    df=pd.DataFrame(rs.fetchmany(size=5))
    df.columns=rs.keys()
print(df)
```

# INNER JOIN

```python
# Apply join in between two tables created out of single csv file.
# To select we can clearly use pandas
innerJoin = pd.read_sql_query('Select diabetes_full_database_table.Glucose, diabetes_full_database_table.Insulin, diabetes_full_database_table.Age, diabetes_selected_database_table.Age from diabetes_full_database_table INNER JOIN diabetes_selected_database_table on diabetes_full_database_table.Age = diabetes_selected_database_table.Age ', engine)
print(innerJoin)
```
